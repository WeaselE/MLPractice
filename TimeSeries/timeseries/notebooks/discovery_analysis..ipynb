```python
# from sklearn.
import pandas as pd

df = pd.read_csv('goog.csv')
df.head()
```

```python
df.describe()
```

```python
import seaborn as sns

sns.lineplot(data=df)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
fig.autofmt_xdate()

sns.swarmplot(data=df, y='Open', hue='Close')
```

```python
sns.boxenplot(data=df)
```

```python
plt.figure(figsize=(20,16))
#row 1
plt.subplot(3,2, 1)
plt.title('Open vs. Close')
sns.scatterplot(data=df, x='Open', y='Close', hue='Open')

plt.subplot(3,2, 2)
plt.title('Open vs. High')
sns.scatterplot(data=df, x='Open', y='Close', hue='Open')

#row 2
plt.subplot(3,2, 3)
plt.title('Open vs. Low')
sns.scatterplot(data=df, x='Open', y='Low', hue='Open')

plt.subplot(3,2, 4)
plt.title('Close vs. High')
sns.scatterplot(data=df, x='Close', y='High', hue='Close')

#row 3
plt.subplot(3,2, 5)
plt.title('Close vs. Low')
sns.scatterplot(data=df, x='Close', y='Low', hue='Close')

plt.subplot(3,2, 6)
plt.title('High vs. Low')
sns.scatterplot(data=df, x='High', y='Low', hue='High')
```
